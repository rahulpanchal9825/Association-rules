```python
import pandas as pd
from mlxtend.frequent_patterns import apriori,association_rules
```

```python
#Q-1
book=pd.read_csv("E:\\DATA SCIENCE\\LMS\\ASSIGNMENT\\MY ASSIGNMENT\\Accosiation\\book.csv")
book
```

```python
frequent_items=apriori(book,min_support=0.1,use_colnames=True)
frequent_items
```

```python
rules = association_rules(frequent_items,metric='lift',min_threshold=0.7)
rules
```

```python
rules.sort_values('lift',ascending = False)[0:20]
```

```python
rules[rules.lift>1]
```

```python
rules[rules.lift>1]
```
